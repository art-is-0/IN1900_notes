```python
import numpy as np
import math
def exp_approx(x, n):
    s = 0
    for k in range(n+1):
        s += x**k / math.factorial(k)
    return s 


for i in range(10):
    x = exp_approx(1, i)
    print(x, np.exp(1))

def test_exp_approx():
    tol = 1e-4
    n = 10
    x = 1

    expected = np.exp(x)
    computed = exp_approx(x, n)

    msg = f"Expected {expected}, computed {computed}"
    assert abs(expected - computed) < tol, msg
    print("Test successful")

test_exp_approx()
```

```python
# v1 = [0,0]; v2 = [1,0]; v3 = [0,2]
# vertices = [v1, v2, v3]

def triangle_area(vertices):
    x1, y1 = vertices[0]
    x2, y2 = vertices[1]
    x3, y3 = vertices[2]

    return 0.5 * abs(x2*y3 - x3*y2 - x1*y3 + x3*y1 + x1*y2 - x2*y1)


def test_triangle_area():
    v1 = [0,0]; v2 = [1,0]; v3 = [0,2]
    vertices = [v1, v2, v3]
    expected = 1
    computed = triangle_area(vertices)
    tol = 1e-10
    success = abs(expected - computed) < tol
    assert success
    print("Working")

test_triangle_area()
```

```python
# a)
class Diff:
    def __init__(self, f, h=1e-4):
        self.f = f
        self.h = h

    def __call__(self, x):
        f, h = self.f, self.h
        df = (f(x+h) - f(x)) / h
        return df
    
# b)
import numpy as np
import matplotlib.pyplot as plt

f = lambda x : np.sin(2*np.pi*x)
x = np.linspace(0, 1, 100)
diff = Diff(f)
plt.plot(x, Diff(f)(x))
plt.show()
```

```python
class Vec2D:
    def __init__(self, components):
        self.x = components[0]
        self.y = components[1]

    def __add__(self, other):
        return Vec2D([self.x + other.x, self.y + other.y])
    
    def __str__(self):
        return f"({self.x},{self.y})"
    
    def __mul__(self, other):
        return self.x * other.x + self.y * other.y
    

x = Vec2D([1,2])
y = Vec2D([4,4])
z = x + y
print(z)
print(type(z))
print(x * y)
```

```python
def bogdanox(N, k, x0, y0):
    x = np.zeros(N)
    y = np.zeros(N)

    x[0] = x0
    y[0] = y0

    for i in range(0, N-1):
        y[i+1] = y[i] + k * x[i]*(x[i] - 1)
        x[i+1] = x[i] + y[i+1]

    return (x, y)

# artisb@uio.no
N = 1000
x0 = y0 = 0.1
k_list = [1.2, 2]

for k in k_list:
    x, y = bogdanox(N, k, x0, y0)

    plt.plot(x, y)

plt.show()
```
